## On titanic daTASET predicting the survived with decision tree

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("titanic.csv")
df.head(10)
```

```python
new_df= df.drop(['Name','SibSp','Ticket','Cabin','Embarked'],axis='columns')
```

```python
new_df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder # Lebel Encoding for converting text to data
```

```python
le_sex=LabelEncoder()
```

```python
new_df['Sex']=le_sex.fit_transform(new_df['Sex'])
```

```python
#new_df.drop(['PassengerId'],inplace=True,axis='columns')
new_df.drop(['Parch'],inplace=True,axis='columns')


new_df.head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_df[['Pclass','Sex','Age','Fare']],new_df['Survived'],test_size=.2)
```

```python
model= DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```
