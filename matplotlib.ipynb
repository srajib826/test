## Bar Graph

```python
import matplotlib.pyplot as plt
```

```python
x= ["python","c","c++","java","ruby"]
y=[85,50,65,80,74]
plt.xlabel("languages")
plt.ylabel("popularity")
c=["y","b","r","o","m"]
plt.bar(x,y,width=0.4,color="c",edgecolor="black",linewidth=2,linestyle="--",alpha=.8)
plt.show()
```

```python
import numpy as np
x= ["python","c","c++","java","ruby"]
y=[85,50,65,80,74]
z=[55,84,23,45,35]
e=[10,12,20,14,30]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
p2=[i+width for i in p1]


plt.xlabel("languages")
plt.ylabel("popularity")


c=["y","b","r","o","m"]
plt.bar(p,y,width,color="c")
plt.bar(p1,z,width,color="b")
plt.bar(p2,e,width,color="black")
plt.xticks(p+width,x,rotation=20)
plt.show()
```

# Scatter plot

```python
x=np.random.randint(1,100,size=(15,))
y=np.random.randint(1,100,size=15)
c=np.random.randint(1,100,size=15)
plt.scatter(x,y,c=c,cmap="viridis")
plt.colorbar()
plt.show()
```
